# Which NYC schools have the **best math results**? 

- At least 80% of maximum possible score (>= 80% of 800)
- Save results to DF `best_math_schools`, include "school_name", "average_math", descending order

```python
# Create filter equations
best_math_results = schools['average_math'] >= (0.8 * 800) # >= 640

```

```python
# Filter the data
high_math_avg = schools.loc[best_math_results, ["school_name", "average_math"]]
high_math_avg_sorted = high_math_avg.sort_values("average_math", ascending=False)

# Save to dataframe
best_math_schools = pd.DataFrame(high_math_avg_sorted)
print(type(best_math_schools))
```

# What are the **top 10 performing schools** based on the **combined SAT scores**? 

```python
# Total SAT
columns_to_sum = ["average_math", "average_reading", "average_writing"]
schools["total_SAT"] = schools[columns_to_sum].sum(axis=1)

# Sort Values by total_SAT 
schools_sorted_by_total_SAT = schools.sort_values(by="total_SAT", ascending=False) 

# limit to top 10 and save to dataframe
top_10_schools = schools_sorted_by_total_SAT.iloc[:10]
top_10_schools = top_10_schools.loc[:, ["school_name", "total_SAT"]]
print(top_10_schools)
```

# **Which single borough has the largest standard deviation in the combined SAT score?**

```python
# Groupby Borough
borough_summary_stats = schools.groupby('borough').agg(
    num_schools = ('school_name', 'count'),
    average_SAT = ('total_SAT', 'mean'), 
    std_SAT = ('total_SAT', 'std'))
borough_summary_stats = borough_summary_stats.round({
    'num_schools': 0, 
    'average_SAT': 2, 
    'std_SAT': 2})

largest_std = borough_summary_stats["std_SAT"].max()
filter_largest_std = borough_summary_stats["std_SAT"] == largest_std
largest_std_dev = borough_summary_stats.loc[filter_largest_std, :]

largest_std_dev
```
